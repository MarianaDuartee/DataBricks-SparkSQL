```

spark.read.format("csv").option("header","true").option("inferSchema","true").option("delimiter",";").load("/FileStore/tables/arquivo_geral.csv").createOrReplaceTempView("tabela_covid")
```

```
%sql

DESCRIBE tabela_covid
```

Retornar a soma de casosNovos por regiao, ordenando de acordo com a soma

```
%sql

SELECT SUM(casosNovos) AS soma_casosNovos,regiao
FROM tabela_covid
GROUP BY regiao
ORDER BY soma_casosNovos
```

Retornar a soma de casos novos, o valor minimo de casosNovos, o valor maximo de casosNovos  e a media de casosNovos, ordenando por regiao

```
%sql

SELECT regiao,SUM(casosNovos) as soma, MIN(casosNovos) as min, MAX(casosNovos) as max, AVG(casosNovos) as media
FROM tabela_covid
GROUP BY regiao
ORDER BY regiao
```
